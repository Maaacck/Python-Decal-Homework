```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
 # 1a
dataFrame = pd.read_csv("GlobalLandTemperaturesByState.csv")
dataFrame = dataFrame.loc[:, ['dt', 'AverageTemperature', 'State']]
dataFrame = dataFrame.loc[(dataFrame['dt'] >= '2000-01-01')]
dataFrame.loc[(dataFrame['State'] == 'Nebraska') | (dataFrame['State'] == 'South Dakota') | (dataFrame['State'] == 'Wyoming')]
```

```python
# 1b
dataFrameTemp = dataFrame.groupby('dt')['AverageTemperature'].mean().reset_index()
dataFrameTemp
```

```python
# 1c
plt.figure(figsize = (200, 15))
plt.scatter(dataFrameTemp['dt'], dataFrameTemp['AverageTemperature'])
plt.title("Date and Average Temperature")
plt.xlabel("Date")
plt.ylabel("Average Temperature")
plt.show()
```

```python
# 1d
dataFrameTempInt = dataFrameTemp
dataFrameTempInt['dt'] = pd.to_datetime(dataFrameTempInt['dt']).dt.strftime('%Y%m%d').astype(int) 
dataFrameTempInt
```

```python
plt.figure(figsize = (15, 5))
plt.scatter(dataFrameTempInt['dt'], dataFrameTemp['AverageTemperature'])
plt.title("Date and Average Temperature")
plt.xlabel("Date")
plt.ylabel("Average Temperature")
plt.show()
```

```python
# 1e 
def equation(x, a, b, c, d):
    return a * np.cos(b * x + c) + d

parameterGuess = np.array([1, 1, 1, 1])
```

```python
plt.figure(figsize = (15, 5))
plt.scatter(dataFrameTempInt['dt'], dataFrameTemp['AverageTemperature'])
plt.title("Date and Average Temperature")
plt.xlabel("Date")
plt.ylabel("Average Temperature")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

x_data = dataFrameTempInt['dt']
y_data = dataFrameTemp['AverageTemperature']

# Initial guess for the parameters
initial_guess = [10, 0.2, 1, 0]
#np.array([10, 0.5, -1, 1])

# Perform the curve fitting
params, params_covariance = curve_fit(equation, x_data, y_data, p0=initial_guess)

# Define the fitted model with the obtained parameters
def fitted_model(x):
    return equation(x, *params)

# Generate x values for the fitted curve
x_fit = np.linspace(x_data.min(), x_data.max(), 100)
y_fit = fitted_model(x_fit)

# Plot the data points
plt.scatter(x_data, y_data, label='Data Points', color='red')

# Plot the fitted curve
plt.plot(x_fit, y_fit, label='Fitted Curve', color='blue')

# Add labels and legend
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Curve Fitting')
plt.legend()

# Show the plot
plt.show()
```
